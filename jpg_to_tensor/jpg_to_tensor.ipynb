## Convert .jpg file to PyTorch Tensor
## Video: https://www.youtube.com/watch?v=UxWhqI8BvCA

In this tutorial we will create a PyTorch Tensor from a .jpg image.

```python
from skimage.transform import resize
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import numpy as np
import torch
import imageio
```

# Loading image from .jpg file

```python
picture = imageio.imread('christmas.jpg')
print("NumPy: ", type(picture))
print("Shape of Array: ", picture.shape)
print("dtype of Array: ", picture.dtype)
```

# Showing the .jpg file as a array with 3 RGB values for every pixel

```python
print(picture)
```

# Resize the Array to fit the Input of the Neural Network

```python
picture = resize(picture, (100,100), mode='constant', anti_aliasing=True)
print(picture)
```

# Show the image, just to check the result

```python
plt.imshow(picture,cmap="gray")
plt.show()
```

# Convert image to Grayscale to reduce the color information (and speed up training later)

```python
picture = rgb2gray(picture)
print(picture)
```

# Show the image, just to check the result

```python
plt.imshow(picture,cmap="gray")
plt.show()
```

<img src="20181101 Image in Grayscale Icon.JPG">

# Convert Array to Tensor with shape (100,100)

```python
Tensor = torch.tensor(picture)
print(Tensor)
print(Tensor.shape)
```

# Flatten the Tensor to use it as Input for Neural Network with shape (10000)

```python
Tensor = Tensor.flatten()
print(Tensor)
print(Tensor.shape)
```
